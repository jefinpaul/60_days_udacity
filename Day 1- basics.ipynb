```python
from skimage import io
photo=io.imread('phont.png')
type(photo)
```

```python
#rows by columns by rgb 
photo.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
#start stop step
plt.imshow(photo[::-1])
```

```python
#columns now are in reverse order so we get a mirror image
plt.imshow(photo[:,::-1])
```

```python
#crop the image
plt.imshow(photo[50:500,180:450])
```

```python
#half the size of the image
plt.imshow(photo[::2,::2])
```
